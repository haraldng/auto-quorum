```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.array([0.125, 0.25, 0.5, 1])
n3 = np.array([0.1, 4.5, 7.6, 21.3])
n5 = np.array([3.2, 6.9, 18.6, 41.4])

# Create a figure and axis
fig, ax = plt.subplots()

# Set the bar width
bar_width = 0.35

# Define the positions for the bars
r1 = np.arange(len(x))  # Positions for n3 bars
r2 = [pos + bar_width for pos in r1]  # Positions for n5 bars

# Create the bars for n3 and n5
ax.bar(r1, n3,  width=bar_width, label='N=3')
ax.bar(r2, n5,  width=bar_width, label='N=5')

# Add labels and title
ax.set_xlabel('Flush duration (ms)')
ax.set_ylabel('Overhead on commit latency (%)')
ax.set_title('IO Overhead on Commit Latency')
ax.set_xticks([r + bar_width / 2 for r in r1])  # Centering the ticks between bars
ax.set_xticklabels(x)

# Add legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.savefig("./overhead.pdf", format='pdf')
plt.show()
```
